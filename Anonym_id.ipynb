```python
LENGTH_ID_USER = 10

# Retourner une chaîne de caractères aléatoire
def generate_new_id_user():
    letters = string.ascii_letters
    return ''.join(random.choice(letters) for _ in range(LENGTH_ID_USER))

# fonctions de hachage sha256 pour id_user
def sha_256():
    string = generate_new_id_user()
    return hashlib.sha256(string.encode()).hexdigest()
```

```python
df_id_user = pd.DataFrame(columns=['id_user'] + list(range(11))) #11 est numero de weekend
df_id_user['id_user'] = df['id_user'].drop_duplicates()
```

```python
# Appliquer sha256 pour les différents id_user pour chaque semaine
def set_df_id_user_week():
    for idx in df_id_user.index:
        for m in range(11): #11 est numero de weekend
            df_id_user.at[idx, m] =  sha_256()
```
