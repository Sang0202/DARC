```python
df2=df2[df2['gid']!='DEL']

df2['week']=pd.to_datetime(df2['datetime']).dt.isocalendar().week

df1['week']=pd.to_datetime(df1['datetime']).dt.isocalendar().week
```

```python
df1['datetime'] = df1['datetime'].str.slice(stop=16)
df1['datetime'] = pd.to_datetime(df1['datetime'], format='%Y-%m-%d %H:%M')
```

```python
df2['datetime'] = df2['datetime'].str.slice(stop=16)
df2['datetime'] = pd.to_datetime(df2['datetime'], format='%Y-%m-%d %H:%M')
```

```python
import numpy as np
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]
```

```python
weeks=df1['week'].unique()
big={}
for w in weeks:
    print(f'week {w}')
    j=0
    dftmp1=df1[df1['week']==w]
    dftmp2=df2[df2['week']==w]
    
    dftmp1["week_day"] = dftmp1["datetime"].dt.isocalendar().day    
    dftmp2["week_day"] = dftmp2["datetime"].dt.isocalendar().day    
    
    name="2015-"+str(w)
    
#     dftmp1=dftmp1.groupby('id', as_index=False)['lat'].mean().sort_values('lat')
#     dftmp2=dftmp2.groupby('gid', as_index=False)['lat'].mean().sort_values('lat')
    
    dftmp1["week_day"] = dftmp1["datetime"].dt.isocalendar().day
    
    dftmp1=dftmp1.groupby('id', as_index=False)['week_day'].mean().sort_values('week_day')
    dftmp2=dftmp2.groupby('gid', as_index=False)['week_day'].mean().sort_values('week_day')
    
    print(dftmp1)
    print(dftmp2)
    
    for i in dftmp2['gid'].to_numpy():
        origId=str(dftmp1.iloc[j,0])
        nearest_value=find_nearest(dftmp2['week_day'].to_numpy(),dftmp1.iloc[j,1] )
        pred=dftmp2[dftmp2['week_day']==nearest_value].iloc[0,0]
        if origId in big:
            big[origId][name]=[str(pred)]
        else:
            dictmp={}
            dictmp[name]=[str(pred)]
            tmp_dict={
                origId:dictmp
            }
        big.update(tmp_dict)
        j+=1
```

```python
with open('549.json', 'w') as f:
    json.dump(big, f)
```
